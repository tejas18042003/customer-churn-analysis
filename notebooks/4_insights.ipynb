## Key Insights

- Customers on month-to-month contracts have the highest churn rate
- Short-tenure customers are significantly more likely to churn compared to longer tenures
- Higher monthly charges are associated with increased churn risk
- Long-term contracts act as a strong retention mechanism
- Customers opting for fiber optics have a higher chance to churn 
  

### Evaluation Focus
- **Recall prioritized over accuracy**
- Reason: Missing a churn is more costly than giving incentive to loyal customers

## Business Recommendations

1. Encourage long-term contracts through incentives
2. Encourage customers to move from month-to-month to long-term contracts, as month-to-month has highest churn rate
3. Target high-risk customers early in their lifecycle
4. Monitor high monthly charge customers for churn signals
5. Fiber optics when opted for as internet plan has high churn rate indicating an underlying issue which needs to be investigated 
6. Use churn predictions to drive proactive retention campaigns using the chosen model from the model fitting section due to good RECALL(0.8) which is perfect for churn
